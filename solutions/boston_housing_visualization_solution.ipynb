# Boston Housing Visualization

## Setup

```python
# Python ≥3.5 is required
import sys
import pandas as pd
assert sys.version_info >= (3, 5)
from IPython.display import display, HTML

sys.path.append("..")

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "end_to_end_project"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)


# Ignore useless warnings (see SciPy issue #5998)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
```

Select the Input File

```python
input_file = "../data/Boston_Housing_Data.csv"
```

Create a DataFrame using an ifered schema 

```python
df = pd.read_csv(input_file, sep=';')
print(df.info())
```

## Plot the data

```python
display (df)
```

### Box Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 25))
plt.suptitle("Boxplots of Boston Housing", fontsize=18,y=1.00)
plt.subplots_adjust(hspace=0.5)

for n, col in enumerate(df.select_dtypes(include=np.number).columns):
    ax = plt.subplot(10, 2, n+1)
    df[[col]].boxplot(ax=ax,sym='r+')
    

save_fig("attribute_box_plots")
plt.show()
```

### Histograms

```python
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20,15))
save_fig("attribute_histogram_plots")
plt.show()
```

### Density Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
df[["CRIM", "ZN", "INDUS"]].plot.kde(figsize=(20,15))
save_fig("attribute_density_plots")
plt.show()
```

### Scatter Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
ax = df.loc[df["CAT"]==0 ,["CRIM", "ZN", "INDUS"]].plot.scatter(x='CRIM', y='INDUS', color='DarkBlue', label='CAT 0')
df.loc[df["CAT"]==0 ,["CRIM", "ZN", "INDUS"]].plot.scatter(x='ZN', y='INDUS', color='DarkGreen', label='CAT 1', ax=ax)
save_fig("scatter_plots")
plt.show()
```

### Scatter matrix

```python
%matplotlib inline
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(df[["CRIM", "ZN", "INDUS","RM", "AGE", "DIS"]], alpha=0.2, figsize=(30, 30), diagonal='kde')
save_fig("scatter_matrix")
plt.show()
```

### Bubble Charts

```python
%matplotlib inline
import matplotlib.pyplot as plt
dfPandas= df[["CRIM", "ZN", "INDUS","MEDV"]]
dfPandas.plot.scatter(x='ZN', y='INDUS', s=dfPandas['MEDV'] * 1, color='DarkBlue')
save_fig("bubble_plots")
plt.show()
```

### Parallel coordinates

```python
%matplotlib inline
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
parallel_coordinates(df[["CRIM", "ZN", "INDUS","CAT"]], 'CAT',color=["r","b"])
save_fig("paralle_coordinates_plots")
plt.show()
```
