# Iris Standardization / Normalization

```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from IPython.display import display, HTML

assert sys.version_info >= (3, 5)
sys.path.append("..")

input_file = "../data/iris.data"
```

```python
# load data file.
# create a DataFrame using an ifered Schema 
df = pd.read_csv(input_file, names=["sepal length","sepal width","petal length","petal width", "label"]) 
num_column_names = df.select_dtypes(include=np.number).columns.tolist()
display(df.describe())
```

## Standardization

```python
print (num_column_names)
scaler = StandardScaler().set_output(transform="pandas")

df_stand = scaler.fit_transform(df[num_column_names])
display(df_stand)
display(df_stand.describe())
```

## Min / Max Normalization


```python
scaler = MinMaxScaler().set_output(transform="pandas")
df_min_max = scaler.fit_transform(df[num_column_names])
display(df_min_max)
display(df_min_max.describe())
```
